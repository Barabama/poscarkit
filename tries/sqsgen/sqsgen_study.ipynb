```python
from sqsgenerator import parse_config, optimize, write
from sqsgenerator.core import LogLevel


with open("CoNiV-1a-Au.json") as f:
    config = parse_config(f.read())

pack = optimize(config, level=LogLevel.info)
best = pack.best()

print(f"best structure: {best.structure()}")
print(f"best sublattices: {best.sublattices()}")
print(f"best objective : {best.objective}")

write(best.structure(), "sqs-CoNiV-1a.vasp")
```
